```python
import os
from dotenv import load_dotenv
load_dotenv()
from langchain_groq import ChatGroq
```

```python
api_key=os.getenv("groq_api_key")
api_key
```

```python
llm=ChatGroq(model="llama-3.3-70b-versatile",api_key=api_key)
llm
```

```python
messages = [  
("system", "You are a creative story writer you are expert in story writing and your creativity is at peaks you generate different mindblowing beautiful stories for all age groups."),  
("human", "Write a story for 1 year child giving final moral of ethics and values."),  
]  
llm.invoke(messages)
```

```python
from langchain.prompts import ChatPromptTemplate
from langchain.schema import HumanMessage,SystemMessage


age_group=input("enter age group(infant/children):") 
story_tone=input("enter story tone(soft/hard):")
moral_theme=input("enter your moral theme(love/affection/sharing/your wish):")
language=input("enter the language:")

prompt=ChatPromptTemplate.from_messages([
    SystemMessage(content="You are a creative story writer you are expert in story writing and your creativity is at peaks you generate different mindblowing beautiful stories for all age groups."),
    HumanMessage(content=f"generate a story for  {age_group}.the story should be in {story_tone} tone, the theme of story is {moral_theme}, the story shoul be in {language}"),
]) 
messages = prompt.format_messages()
response=llm.invoke(messages)
response.content
```
